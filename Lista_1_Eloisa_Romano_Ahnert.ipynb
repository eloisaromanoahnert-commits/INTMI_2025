```python
# Solicita ao usuário que digite "S" para continuar ou "N" para sair
resposta = input('Digite "S" para continuar ou "N" para sair: ').strip().upper()

# Verifica a resposta e mostra a ação correspondente
if resposta == "S":
    print("Você escolheu continuar!")
elif resposta == "N":
    print("Você escolheu sair. Até logo!")
else:
    print("Opção inválida! Digite apenas 'S' ou 'N'.")
```

```python
# Solicita um número ao usuário
numero = float(input("Digite um número: "))

# Verifica se o número é negativo, zero ou positivo
if numero > 0:
    print("O número é positivo.")
elif numero == 0:
    print("O número é zero.")
else:
    print("O número é negativo.")
```

```python
# Solicita um número de 1 a 7 ao usuário
numero = int(input("Digite um número de 1 a 7: "))

# Verifica o número e exibe o dia correspondente
if numero == 1:
    print("Domingo")
elif numero == 2:
    print("Segunda-feira")
elif numero == 3:
    print("Terça-feira")
elif numero == 4:
    print("Quarta-feira")
elif numero == 5:
    print("Quinta-feira")
elif numero == 6:
    print("Sexta-feira")
elif numero == 7:
    print("Sábado")
else:
    print("Número inválido! Digite um valor entre 1 e 7.")
```

```python
# Solicita o ano ao usuário
ano = int(input("Digite um ano: "))

# Verifica se o ano é bissexto
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print("O ano é bissexto.")
else:
    print("O ano não é bissexto.")
```

```python
# Solicita os dados ao usuário
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_hora = float(input("Digite o valor recebido por hora: "))

# Verifica se houve horas extras
if horas_trabalhadas <= 40:
    salario = horas_trabalhadas * valor_hora
else:
    horas_normais = 40
    horas_extras = horas_trabalhadas - 40
    salario = (horas_normais * valor_hora) + (horas_extras * valor_hora * 1.5)

# Exibe o resultado
print(f"Salário total: R$ {salario:.2f}")
```
