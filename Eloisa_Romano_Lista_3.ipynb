```python
import random

def sortear_mega_sena():
    """Simula um sorteio de 6 números da Mega-Sena sem repetição."""

    VALOR_MIN = 1
    VALOR_MAX = 60
    QUANTIDADE_NUMEROS = 6

    numeros_sorteados = random.sample(range(VALOR_MIN, VALOR_MAX + 1), QUANTIDADE_NUMEROS)

    numeros_sorteados.sort()

    return numeros_sorteados
resultado_sorteio = sortear_mega_sena()
print(f"--- Sorteio Mega-Sena ---")
print(f"Os 6 números sorteados foram: {resultado_sorteio}")
```

```python
N = 3
matriz = []
is_identidade = True

print(f"Por favor, digite {N*N} números inteiros para preencher a matriz {N}x{N}:")

for i in range(N):
    linha = []
    for j in range(N):
        while True:
            try:
                valor = int(input(f"Digite o valor para a posição ({i+1},{j+1}): "))
                linha.append(valor)
                break
            except ValueError:
                print("Entrada inválida. Digite um número inteiro.")
    matriz.append(linha)

print("\nMatriz inserida:")
for linha in matriz:
    print(linha)

for i in range(N):
    for j in range(N):
        if i == j:
            # Se for a diagonal principal, o valor deve ser 1
            if matriz[i][j] != 1:
                is_identidade = False
                break
        else:

            if matriz[i][j] != 0:
                is_identidade = False
                break
    if not is_identidade:
        break

print("-" * 30)
if is_identidade:
    print("A matriz informada é uma MATRIZ IDENTIDADE.")
else:
    print("A matriz informada NÃO é uma matriz identidade.")
print("-" * 30)
```

```python
N = 3
matriz_A = []
matriz_B = []
matriz_soma = []
matriz_diferenca = []

def ler_matriz(nome_matriz, tamanho):
    """Função auxiliar para ler uma matriz do usuário."""
    matriz_temp = []
    print(f"\n--- Preencher Matriz {nome_matriz} ({tamanho}x{tamanho}) ---")
    for i in range(tamanho):
        linha = []
        for j in range(tamanho):
            while True:
                try:
                    valor = int(input(f"Digite A[{i+1}][{j+1}] para {nome_matriz}: "))
                    linha.append(valor)
                    break
                except ValueError:
                    print("Entrada inválida. Digite um número inteiro.")
        matriz_temp.append(linha)
    return matriz_temp
matriz_A = ler_matriz("A", N)
matriz_B = ler_matriz("B", N)
for i in range(N):
    linha_soma = []
    linha_diferenca = []
    for j in range(N):
        soma = matriz_A[i][j] + matriz_B[i][j]
        diferenca = matriz_A[i][j] - matriz_B[i][j]
        linha_soma.append(soma)
        linha_diferenca.append(diferenca)
    matriz_soma.append(linha_soma)
    matriz_diferenca.append(linha_diferenca)
print("\n--- Resultados ---")
print("Matriz A:")
for linha in matriz_A: print(linha)
print("\nMatriz B:")
for linha in matriz_B: print(linha)
print("\nMatriz Soma (A + B):")
for linha in matriz_soma: print(linha)
print("\nMatriz Diferença (A - B):")
for linha in matriz_diferenca: print(linha)
```

```python
NUM_ELEMENTOS = 10

def ler_conjunto(nome_conjunto, tamanho):
    """Função auxiliar para ler um conjunto de números do usuário."""
    conjunto_temp = set()
    print(f"\n--- Preencher Conjunto {nome_conjunto} ({tamanho} elementos únicos) ---")
    while len(conjunto_temp) < tamanho:
        try:
            numero = float(input(f"Digite o {len(conjunto_temp) + 1}º número para {nome_conjunto}: "))
            if numero in conjunto_temp:
                print(f"O número {numero} já está no conjunto. Digite um número diferente.")
            else:
                conjunto_temp.add(numero)
        except ValueError:
            print("Entrada inválida. Por favor, digite um número válido.")
    return conjunto_temp
conjunto_A = ler_conjunto("A", NUM_ELEMENTOS)
conjunto_B = ler_conjunto("B", NUM_ELEMENTOS)

print(f"\nConjunto A: {conjunto_A}")
print(f"Conjunto B: {conjunto_B}")
intersecao = conjunto_A & conjunto_B
uniao = conjunto_A | conjunto_B
diferenca_A_menos_B = conjunto_A - conjunto_B
diferenca_B_menos_A = conjunto_B - conjunto_A
print("\n--- Resultados das Operações ---")
print(f"Interseção (A ∩ B): {intersecao}")
print(f"União (A ∪ B): {uniao}")
print(f"Diferença (A - B): {diferenca_A_menos_B}")
print(f"Diferença (B - A): {diferenca_B_menos_A}")
```
