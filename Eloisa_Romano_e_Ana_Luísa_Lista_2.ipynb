```python
soma_total = 0

print("Calculadora de Soma. Digite 0 para finalizar e ver o resultado.")

while True:
    numero = float(input("Digite um número: "))
    if numero == 0:
        break  # Interrompe o loop quando 0 é digitado
    soma_total += numero

print(f"A soma total dos números digitados é: {soma_total}")
```

```python
soma = 0
numero = 1

print("Iniciando a simulação. Números utilizados:")

while soma <= 100:
    print(f"Usando o número: {numero}")
    soma += numero
    numero += 1

print(f"Fim da simulação. A soma total ultrapassou 100, chegando a {soma}.")
```

```python
soma_idades = 0
contador_idades = 0

print("Pesquisa de idades. Digite um número negativo para finalizar.")

while True:
    idade = int(input("Digite uma idade: "))

    if idade < 0:
        break  # Sai do loop se a idade for negativa

    soma_idades += idade
    contador_idades += 1

if contador_idades > 0:
    media = soma_idades / contador_idades
    print(f"Total de idades registradas: {contador_idades}")
    print(f"Soma total das idades: {soma_idades}")
    print(f"Média das idades: {media:.1f}")
else:
    print("Nenhuma idade válida foi digitada para calcular a média.")
```

```python
numero_inicial = int(input("Digite um número inicial para a contagem regressiva: "))

if numero_inicial < 0:
    print("O número inicial deve ser positivo ou zero.")
else:
    print(f"Iniciando contagem regressiva a partir de {numero_inicial}:")
    contador = numero_inicial
    while contador >= 0:
        print(contador)
        contador -= 1
    print("Contagem regressiva concluída!")
```

```python
contador_maiores_18 = 0

print("Digite 5 idades para a análise:")

# O loop executa 5 vezes (de 0 a 4)
for i in range(5):
    idade = int(input(f"Idade {i+1}: "))
    if idade > 18:
        contador_maiores_18 += 1

print(f"Das 5 idades digitadas, {contador_maiores_18} são maiores que 18 anos.")
```

```python
maior_numero = float('-inf') # Inicializa com um valor muito baixo para garantir que o primeiro número seja maior

print("Digite 7 números para determinar o maior:")

for i in range(7):
    numero = float(input(f"Número {i+1}: "))
    if numero > maior_numero:
        maior_numero = numero

print(f"O maior número digitado foi: {maior_numero}")
```

```python
nome_mais_longo = ""
tamanho_maximo = 0

print("Digite os nomes de 4 tarefas:")

for i in range(4):
    nome_tarefa = input(f"Tarefa {i+1}: ")
    if len(nome_tarefa) > tamanho_maximo:
        tamanho_maximo = len(nome_tarefa)
        nome_mais_longo = nome_tarefa

print(f"A tarefa com o nome mais longo é: '{nome_mais_longo}'")
```

```python
positivos = 0
negativos = 0
zeros = 0

print("Digite 10 números para a análise:")

for i in range(10):
    numero = float(input(f"Número {i+1}: "))
    if numero > 0:
        positivos += 1
    elif numero < 0:
        negativos += 1
    else:
        zeros += 1

print(f"Análise concluída:")
print(f"Positivos: {positivos}")
print(f"Negativos: {negativos}")
print(f"Zeros: {zeros}")
```

```python
print("Digite 6 nomes:")

for i in range(6):
    nome = input(f"Nome {i+1}: ")
    # Verifica se a primeira letra (em minúsculo) é 'a' ou 'c'
    if nome.lower().startswith('a') or nome.lower().startswith('c'):
        print(f"Nome filtrado: {nome}")
```

```python
soma_valores = 0

print("Caixa Registradora. Digite 999 para fechar a conta.")

while True:
    valor = float(input("Digite o valor do item (ou 999 para encerrar): "))

    if valor == 999:
        break  # Encerra o loop

    soma_valores += valor

print(f"A soma total dos itens da compra é: R$ {soma_valores:.2f}")
```

```python
print("Listando números de 1 a 20 (ignorando múltiplos de 4):")

for numero in range(1, 21):
    if numero % 4 == 0:
        continue  # Pula o restante do bloco e vai para a próxima iteração
    print(numero)
```

```python
SENHA_CORRETA = "python123"
MAX_TENTATIVAS = 3

for tentativa in range(1, MAX_TENTATIVAS + 1):
    senha_digitada = input(f"Tentativa {tentativa}/{MAX_TENTATIVAS}. Digite a senha: ")

    if senha_digitada == SENHA_CORRETA:
        print("Login bem-sucedido! Bem-vindo(a).")
        break  # Sai do loop imediatamente
    elif tentativa == MAX_TENTATIVAS:
        print("Número máximo de tentativas excedido. Acesso bloqueado.")
    else:
        print("Senha incorreta. Tente novamente.")
```

```python
soma_positivos = 0

print("Digite 10 números para somar apenas os valores positivos:")

for i in range(10):
    numero = float(input(f"Número {i+1}: "))

    if numero < 0:
        print(f"Valor negativo {numero} ignorado.")
        continue  # Pula o restante do bloco e vai para a próxima iteração

    soma_positivos += numero

print(f"A soma total dos valores positivos digitados é: {soma_positivos}")
```

```python
soma = 0
numero = 1

print("Iniciando acumulador. Somando números sequenciais:")

while True:
    soma += numero
    print(f"Somando {numero}. Soma atual: {soma}")
    if soma > 200:
        print("Soma ultrapassou 200.")
        break  # Encerra o laço
    numero += 1

print(f"O laço foi encerrado. A soma final é: {soma}")
```
